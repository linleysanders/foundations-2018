```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
response = requests.get('http://www.nytimes.com')
doc = BeautifulSoup(response.text, 'html.parser')
```

# Where are the containers of the stories on the NYT website

```python
stories = doc.find_all(class_='story')
len(stories)
```

```python
for story in stories:
    headline = story.find(class_='story-heading')
    byline = story.find(class_='byline')
    if headline:
        print(headline.text)
    if byline:
        print(byline.text)
```

```python
list_of_stories = []

for story in stories:
    story_dict = {}
    headline = story.find(class_='story-heading')
    byline = story.find(class_='byline')
    if headline:
        story_dict['headline'] = headline.text.strip()
    if byline:
        story_dict['byline'] = byline.text.strip()
    list_of_stories.append(story_dict)

list_of_stories
```

```python
df = pd.DataFrame(list_of_stories)
df.head()
```

```python
df.to_csv("nyt-stories.csv", index=False)
```
